```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from mpl_toolkits import mplot3d
```

```python
df = pd.read_excel('Dataset Pengaruh TPT dan RLS terhadap presentase penduduk miskin2.xlsx')
df
```

```python
df.info()
```

```python
X = df[['Tingkat Pengangguran Terbuka (Persen)', 'Rata-Rata Lama Sekolah Penduduk Umur 15 Tahun ke Atas', 'Upah Minimum Provinsi']]
X.head()
```

```python
y = df['Persentase Penduduk Miskin (Persen)']
y.head()
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)

koef_tpt = regr.coef_[0]
koef_rls = regr.coef_[1]
koef_ump = regr.coef_[2]
konstanta = regr.intercept_
```

```python
y_pred = regr.predict(X)
r_square = r2_score(y, y_pred)
```

```python
print("=================================================================")
print("             PENILAIAN KETEPATAN MODEL (R-SQUARED)               ")
print("=================================================================")
print(f"Nilai R-squared (R²)      : {r_square:.6f}")
print(f"Sumbangan Determinasi (%) : {r_square * 100:.2f}%")
print("-----------------------------------------------------------------")
print("Tafsiran Akademik Projek UAS:")
print(f"Sebanyak {r_square * 100:.2f}% variasi kadar 'Kemiskinan Provinsi'")
print("di Indonesia dapat dijelaskan secara serentak oleh pemandunya iaitu:")
print("1. Tingkat Pengangguran Terbuka (TPT)")
print("2. Rata-Rata Lama Sekolah (RLS)")
print("3. Upah Minimum Provinsi (UMP)")
print(f"sebanyak {(1 - r_square) * 100:.2f}% lagi dijelaskan oleh faktor-faktor")
print("luaran lain yang tidak dimasukkan ke dalam model ini.")
print("=================================================================")
```

```python
def semak_arah_hubungan(pekali, nama_pembolehubah, arah_teori):
    if pekali > 0:
        status = "POSITIF (+)"
        ulasan = f"Setiap peningkatan {nama_pembolehubah} akan cenderung meningkatkan kadar kemiskinan."
    elif pekali < 0:
        status = "NEGATIF (-)"
        ulasan = f"Setiap peningkatan {nama_pembolehubah} akan cenderung menurunkan kadar kemiskinan."
    else:
        status = "NETRAL (0)"
        ulasan = f"Perubahan pada {nama_pembolehubah} tidak memberi kesan kepada kemiskinan."
        
    kesesuaian = "SELARAS TEORI" if status.split()[0].lower() == arah_teori.lower() else "ANOMALI"
    return status, ulasan, kesesuaian

status_tpt, ulasan_tpt, cocok_tpt = semak_arah_hubungan(koef_tpt, "Kadar Pengangguran", "POSITIF")
status_rls, ulasan_rls, cocok_rls = semak_arah_hubungan(koef_rls, "Tempoh Persekolahan", "NEGATIF")
status_ump, ulasan_ump, cocok_ump = semak_arah_hubungan(koef_ump, "Gaji Minimum (UMP)", "NEGATIF")
```

```python
print("=================================================================")
print("               PARAMETER UTAMA MODEL REGRESI                     ")
print("=================================================================")
print(f"Pemalar (a)                    : {konstanta:.6f}")
print(f"Pekali Pengangguran (b1)       : {koef_tpt:.6f} ({status_tpt} - {cocok_tpt})")
print(f"Pekali Tempoh Sekolah (b2)     : {koef_rls:.6f} ({status_rls} - {cocok_rls})")
print(f"Pekali Gaji Minimum UMP (b3)   : {koef_ump:.8f} ({status_ump} - {cocok_ump})")
print("\nPersamaan Matematik Regresi Berganda Baru:")
print(f"Y = {konstanta:.4f} + ({koef_tpt:.4f} * TPT) + ({koef_rls:.4f} * RLS) + ({koef_ump:.8f} * UMP)")
print("\n--- TAFSIRAN IMPAK VARIABEL ---")
print(f"1. Pengangguran: {ulasan_tpt}")
print(f"2. Pendidikan  : {ulasan_rls}")
print(f"3. Gaji (UMP)  : {ulasan_ump}")
print("=================================================================")
```
